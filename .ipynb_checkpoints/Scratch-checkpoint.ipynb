## Proposition: Ogden's model can break polyconvexity if $\lambda_1 \lambda_2 \lambda_3 \neq 1$

```python
F = np.zeros([20,3,3])
for i in range(20):
    F[i,0,0] = 1 + i/20*0.5
    F[i,1,1] = 1
    F[i,2,2] = 1
```

```python
lm,_ = np.linalg.eig(F)
```

```python
mu = 1
a = 0.01
Psi = mu/a*(lm[:,0]**a + lm[:,1]**a + lm[:,2]**a - 3)
```

```python
plt.plot(F[:,0,0],Psi)
plt.plot([F[0,0,0],F[-1,0,0]], [Psi[0], Psi[-1]], c='r')
plt.xlabel('$F_{11}$')
plt.ylabel('$\Psi$')
plt.legend(['Psi','Straight line'])
```

## Proposition: more generally a convex and nondecreasing function of the 
## eigenvalues of a matrix is not necessarily convex wrt the matrix itself

Example 1: $\Psi = exp(\tau_1) + exp(\tau_2) + exp(\tau_3) - 3$

```python
for i in range(1000): #Check 10 different starting points for tau matrix
    tau_voigt = np.random.randn(6)
    indices = np.array([[0,1,2],[1,3,4],[2,4,5]])
    tau = tau_voigt[indices]
    [tau1, tau2, tau3],_ = np.linalg.eig(tau)
    Psi = np.exp(tau1)+np.exp(tau2)+np.exp(tau3)-3
    for j in range(6):
        dx = 0.01
        tau_pvoigt = np.copy(tau_voigt)
        tau_pvoigt[j]+= dx
        tau_p = tau_pvoigt[indices]
        [tau1, tau2, tau3],_ = np.linalg.eig(tau_p)
        Psi_p = np.exp(tau1)+np.exp(tau2)+np.exp(tau3)-3 #f(x2)
        
        tau_pvoigt = np.copy(tau_voigt)
        tau_pvoigt[j]+= dx/2
        tau_p = tau_pvoigt[indices]
        [tau1, tau2, tau3],_ = np.linalg.eig(tau_p)
        Psi_p2 = np.exp(tau1)+np.exp(tau2)+np.exp(tau3)-3 #f((x1+x2)/2)
        
        if 0.5*Psi + 0.5*Psi_p < Psi_p2:
            print('Found nonconvex point')
            print(tau)
            print(j)
            print(Psi, Psi_p2, Psi_p)
            print(0.5*Psi + 0.5*Psi_p)
            break
```

```python
tau_vec = np.array([tau for i in range(20)])
tau_vec[:,0,0] = np.linspace(tau_vec[0,0,0],tau_vec[0,0,0]+1,20)
eig,_ = np.linalg.eig(tau_vec)
Psi = np.exp(eig[:,0])+np.exp(eig[:,1])+np.exp(eig[:,2])-3
```

```python
plt.plot(tau_vec[:,0,0], Psi)
plt.plot([tau_vec[0,0,0], tau_vec[-1,0,0]], [Psi[0], Psi[-1]])
```

Example 2: $\Psi = \tau_1^2 + \tau_2^2 + \tau_3^2$

```python
for i in range(100): #Check 10 different starting points for tau matrix
    tau_voigt = np.random.randn(6)
    indices = np.array([[0,1,2],[1,3,4],[2,4,5]])
    tau = tau_voigt[indices]
    [tau1, tau2, tau3],_ = np.linalg.eig(tau)
    Psi = tau1**2 + tau2**2 + tau3**2
    for j in range(6):
        dx = 0.01
        tau_pvoigt = np.copy(tau_voigt)
        tau_pvoigt[j]+= dx
        tau_p = tau_pvoigt[indices]
        [tau1, tau2, tau3],_ = np.linalg.eig(tau_p)
        Psi_p = tau1**2 + tau2**2 + tau3**2 #f(x2)
        
        tau_pvoigt = np.copy(tau_voigt)
        tau_pvoigt[j]+= dx/2
        tau_p = tau_pvoigt[indices]
        [tau1, tau2, tau3],_ = np.linalg.eig(tau_p)
        Psi_p2 = tau1**2 + tau2**2 + tau3**2 #f((x1+x2)/2)
        
        if 0.5*Psi + 0.5*Psi_p < Psi_p2:
            print('Found nonconvex point')
            print(tau)
            print(j)
            print(Psi, Psi_p2, Psi_p)
            print(0.5*Psi + 0.5*Psi_p)
            break
```

```python
tau_vec = np.array([tau for i in range(20)])
tau_vec[:,0,0] = np.linspace(tau_vec[0,0,0],tau_vec[0,0,0]+1,20)
eig,_ = np.linalg.eig(tau_vec)
Psi = eig[:,0]**2 + eig[:,1]**2 + eig[:,2]**2

plt.plot(tau_vec[:,0,0], Psi)
plt.plot([tau_vec[0,0,0], tau_vec[-1,0,0]], [Psi[0], Psi[-1]])
```

## The 2nd eigenvalue is not a convex/concave function of the matrix elements

```python
tau_voigt = np.random.randn(6)
indices = np.array([[0,1,2],[1,3,4],[2,4,5]])
indices2 = np.array([[0,0], [0,1], [0,2], [1,1], [1,2], [2,2]])
tau = tau_voigt[indices]
eig,_ = np.linalg.eig(tau)
eig = np.flip(np.sort(eig))
eps = 0.01

tauconvexity = []
for i in range(6):
    tau_voigt_m = np.copy(tau_voigt)
    tau_voigt_m[i]-= eps
    tau_m = tau_voigt_m[indices]
    eig_m,_ = np.linalg.eig(tau_m)
    eig_m = np.flip(np.sort(eig_m))
    
    tau_voigt_p = np.copy(tau_voigt)
    tau_voigt_p[i]+= eps
    tau_p = tau_voigt_p[indices]
    eig_p,_ = np.linalg.eig(tau_p)
    eig_p = np.flip(np.sort(eig_p))
    
    d2 = (eig_p - 2*eig + eig_m)/eps**2
    
    tauconvexity.append(np.sign(d2))

```

```python
def highlight_max(s):
    if s.dtype == np.object:
        is_neg = [False for _ in range(s.shape[0])]
    else:
        is_neg = s < 0
    return ['background: red; color:white' 
            if cell else 'background:blue; color:white' 
            for cell in is_neg]

xlabels=['$\mathbf{\\tau}_{11}$','$\mathbf{\\tau}_{12}$','$\mathbf{\\tau}_{13}$','$\mathbf{\\tau}_{22}$','$\mathbf{\\tau}_{23}$','$\mathbf{\\tau}_{33}$']
ylabels = ['$\\tau_{1}$','$\\tau_{2}$','$\\tau_{3}$']
convexity = pd.DataFrame(tauconvexity, index=xlabels, columns = ylabels)
convexity = convexity.T
print('Convexity of principal values of a matrix wrt its elements')
convexity.style.apply(highlight_max)
```

The middle eigenvalue(s) is indeed neither convex nor concave. Now design a function $\Psi = \Psi(\tau_2)$, that is convex in eigenvalues (in this case just $\tau_2$), and show that it is not convex wrt the matrix.

$\Psi = \tau_2^2$

```python
tau_vec = np.array([tau for i in range(20)])
tau_vec[:,0,0] = np.linspace(tau_vec[0,0,0],tau_vec[0,0,0]+1,20)
eig,_ = np.linalg.eig(tau_vec)
eig = np.sort(eig)
Psi = eig[:,1]**2

plt.plot(tau_vec[:,0,0], Psi)
plt.plot([tau_vec[0,0,0], tau_vec[-1,0,0]], [Psi[0], Psi[-1]])
plt.legend(['Strain energy','Straight line'])
plt.xlabel('$\\tau_{11}$')
plt.ylabel('$\Psi$')
pass
```

# Generate training data with Govindjee

### Uniaxial Tension

```python
import numpy as np, matplotlib.pyplot as plt
from material_models import Govindjee
import matplotlib
font = {'size'   : 16}
matplotlib.rc('font', **font)
```

```python
# uniaxial tension 
nsteps = 200
dt = 1
eps_x_unia   = np.zeros(nsteps)
eps_y_unia   = np.zeros(nsteps)
sigma_x_unia = np.zeros(nsteps)
sigma_y_unia = np.zeros(nsteps)
time         = np.zeros(nsteps)
#Initial values for viscous strains
epsilon_v = np.zeros((6))
C_i_inv   = np.eye(3)
for i in range(nsteps):
    if i<20:
        epsilon_x = i/nsteps+1e-6
    else:
        epsilon_x = 0.1
    sigma_y = 0
    sigma_z = 0
    normres = 1
    iter=0
    itermax = 20
    epsilon_y = 0
    epsilon_z = 0
    eta = 10
    while normres>1e-6 and iter<itermax:
        # guess for F
        F = np.array([[1+epsilon_x, 0, 0], [0, 1+epsilon_y, 0], [0, 0, 1+epsilon_z]])
        #print(F)
        # GIVEN F and C_i_inv from previous time, compute stress and new C_i_inv
        sigma, C_i_inv_new, lamb_e = Govindjee(F, C_i_inv, dt)
        res = np.array([sigma[1,1]-sigma_y, sigma[2,2]-sigma_z])
        
        # calculate dres with NR 
        F_py = np.array([[1+epsilon_x, 0, 0], [0, 1+epsilon_y+1e-6, 0], [0, 0, 1+epsilon_z]])
        sigma_py,aux, aux2 = Govindjee(F_py, C_i_inv, dt)
        F_pz = np.array([[1+epsilon_x, 0, 0], [0, 1+epsilon_y, 0], [0, 0, 1+epsilon_z+1e-6]])
        sigma_pz,aux, aux2 = Govindjee(F_pz, C_i_inv, dt)
        dres = np.array([[(sigma_py[1,1]-sigma[1,1])/1e-6,(sigma_py[2,2]-sigma[2,2])/1e-6],\
                        [(sigma_pz[1,1]-sigma[1,1])/1e-6,(sigma_pz[2,2]-sigma[2,2])/1e-6]])
        deps = np.linalg.solve(dres,-res)
        epsilon_y += deps[0]
        epsilon_z += deps[1]
        normres = np.linalg.norm(res)
        iter+=1 

    # update the internal variable at end of iterations 
    C_i_inv = C_i_inv_new
    sigma_x_unia[i] = sigma[0,0]
    sigma_y_unia[i] = sigma[1,1]
    eps_x_unia[i]   = epsilon_x
    eps_y_unia[i]   = epsilon_y    
    time[i] = time[i-1]+dt
```

```python
fig,ax = plt.subplots(1,2,figsize=[14,6])

ax[0].plot(time, eps_x_unia, label='$\epsilon_x$')
ax[0].plot(time, eps_y_unia, label='$\epsilon_y$')
ax[0].legend()
ax[0].set(xlabel='time (s)', ylabel='$\epsilon$')

ax[1].plot(time, sigma_x_unia, label='$\sigma_x$')
ax[1].plot(time, sigma_y_unia, label='$\sigma_y$')
ax[1].legend()
ax[1].set(xlabel='time (s)', ylabel='$\sigma$')

fig.suptitle('Uniaxial tension')
pass
```

### Equibiaxial tension

```python
# Biaxial tension in plane stress
nsteps = 200
dt = 1

eps_x_equi   = np.zeros(nsteps)
eps_y_equi   = np.zeros(nsteps)
sigma_x_equi = np.zeros(nsteps)
sigma_y_equi = np.zeros(nsteps)
time         = np.zeros(nsteps)

# initial condition for viscous strains 
epsilon_v = np.zeros(6)
C_i_inv   = np.eye(3)
for i in range(nsteps):
    if i<10:
        epsilon_x = i/nsteps*2 + 1e-6
        epsilon_y = i/nsteps*2 + 1e-6
    else:
        epsilon_x = 0.1
        epsilon_y = 0.1
    
    sigma_y = 0
    sigma_z = 0
    normres = 1
    iter=0
    itermax = 20
    epsilon_z = 0
    eta = 10
    while normres>1e-6 and iter<itermax:
        # guess for F
        F = np.array([[1+epsilon_x, 0, 0], [0, 1+epsilon_y, 0], [0, 0, 1+epsilon_z]])
        sigma, C_i_inv_new, lamb_e = Govindjee(F, C_i_inv, dt)
        res = sigma[2,2]-sigma_z
        
        # calculate dres with NR 
        F_pz = np.array([[1+epsilon_x, 0, 0], [0, 1+epsilon_y, 0], [0, 0, 1+epsilon_z+1e-6]])
        sigma_pz, aux, aux2 = Govindjee(F_pz, C_i_inv, dt)
        dres = (sigma_pz[2,2]-sigma[2,2])/1e-6
        
        deps = -res/dres
        epsilon_z += deps
        normres = np.linalg.norm(res)
        iter+=1 

    # update the internal variable at end of iterations 
    C_i_inv = C_i_inv_new
    sigma_x_equi[i] = sigma[0,0]
    sigma_y_equi[i] = sigma[1,1]
    eps_x_equi[i]   = epsilon_x
    eps_y_equi[i]   = epsilon_y    
    time[i] = time[i-1]+dt
```

```python
fig,ax = plt.subplots(1,2,figsize=[14,6])

ax[0].plot(time, eps_x_equi, label='$\epsilon_x$')
ax[0].plot(time, eps_y_equi, label='$\epsilon_y$')
ax[0].legend()
ax[0].set(xlabel='time (s)', ylabel='$\epsilon$')

ax[1].plot(time, sigma_x_equi, label='$\sigma_x$')
ax[1].plot(time, sigma_y_equi, label='$\sigma_y$')
ax[1].legend()
ax[1].set(xlabel='time (s)', ylabel='$\sigma$')

fig.suptitle('Equibiaxial tension')
pass
```
