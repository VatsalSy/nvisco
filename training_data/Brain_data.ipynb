```python
def nondecreasing_time(aux):
    aux[:,0] = np.abs(aux[:,0])
    flag1 = True
    mask = np.ones(aux.shape[0], dtype=bool)
    while flag1:
        flag2 = True
        for i in range(1,aux.shape[0]):
            if aux[i,0]<aux[i-1,0]:
                mask[i] = 0
                aux = aux[mask]
                mask = np.ones(aux.shape[0], dtype=bool)
                flag2 = False
                break
        if flag2:
            flag1 = False
    return aux
```
