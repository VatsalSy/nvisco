```python
import numpy as np
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = "serif"
```

```python
fig, ax = plt.subplots(figsize=(10,5))
ax.set(ylim=[-1,0])
x = np.linspace(0,1,50)
f = lambda r, c, x: -c*np.exp(r*x) 
r = 2.0
for c in np.linspace(0.01,0.13,5):
    y = f(r,c,x)
    ax.plot(x,y,'k')

x = np.linspace(0, 0.96,10)
y = np.linspace(-0.96,0,10)
x,y = np.meshgrid(x,y)
x = x.flatten()
y = y.flatten()
u = np.ones_like(x)
v = np.zeros_like(x)
for i in range(v.shape[0]):
    v[i] = -f(r,y[i]/4,x[i])

ax.quiver(x,y,u,v, alpha=0.2)
ax.spines['top'].set_visible(False)
ax.spines['right'].set_visible(False)
fig.savefig('traj_plot.eps')
```
