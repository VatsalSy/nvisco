```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import matplotlib
font = {'size'   : 14}
matplotlib.rc('font', **font)
import pickle
```

```python
s1 = pd.read_csv("manu_bloodclots/Sample_1.csv")
s2 = pd.read_csv("manu_bloodclots/Sample_2.csv")
s4 = pd.read_csv("manu_bloodclots/Sample_4.csv")
s5 = pd.read_csv("manu_bloodclots/Sample_5.csv")
s6 = pd.read_csv("manu_bloodclots/Sample_6.csv")
s7 = pd.read_csv("manu_bloodclots/Sample_7.csv")
s1
```

```python
t = np.array(s1.Time[1:],           dtype=np.float64)
d = np.array(s1.Displacement[1:],   dtype=np.float64)
f = np.array(s1.Force[1:],          dtype=np.float64)

fig, ax = plt.subplots(1,3,figsize=[15,4])
fig.subplots_adjust(wspace=0.3)
ax[0].plot(t,d)
ax[1].plot(t,f)
ax[2].plot(d,f)

ax[0].set(xlabel='Time (s)',            ylabel='Displacement (mm)')
ax[1].set(xlabel='Time (s)',            ylabel='Force (mN)')
ax[2].set(xlabel='Displacement (mm)',   ylabel='Force (mN)'); 
fig.suptitle('Sample 1')
```

```python
minlen = np.min([len(d4), len(d5), len(d6)])
t = np.stack([t4[:minlen],t5[:minlen],t6[:minlen]])
d = np.stack([d4[:minlen],d5[:minlen],d6[:minlen]])
f = np.stack([f4[:minlen],f5[:minlen],f6[:minlen]])

d = d/10 # Displacement -> stretch
```
